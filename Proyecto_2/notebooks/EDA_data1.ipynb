### 3. **Diagnóstico y Selección de Problema:**
* Basándose en el EDA inicial, diagnosticar las principales características y desafíos de cada conjunto de datos.
* Elegir una problemática específica para abordar (regresión, clasificación, clusterización, predicción).
* Justificar la elección del problema y explicar por qué es relevante y desafiante.


**Descripción del problema:**
La tarea seleccionada es la predicción de los cargos médicos de los pacientes basándose en características como la edad, el índice de masa corporal (IMC), el número de hijos, si son fumadores, la región en la que viven, y su género. Este es un problema de regresión porque se trata de predecir un valor continuo: el costo asociado con el seguro de salud, en lugar de clasificar a los pacientes en diferentes categorías.

**Justificación de la elección de la regresión lineal:**
1. **Simplicidad y comprensión:**
Modelo interpretable: La regresión lineal es un modelo estadístico fácil de entender y explicar. Los coeficientes asociados a cada variable (característica) nos indican cómo afecta cada variable independiente (por ejemplo, edad, IMC, fumar, etc.) al valor de la variable dependiente (cargos médicos). Esto puede ser valioso tanto para la interpretación del modelo como para la toma de decisiones basadas en el análisis de los resultados.
Rapidez y eficiencia: La regresión lineal es un modelo simple y eficiente desde el punto de vista computacional. Esto lo convierte en una excelente opción si se busca una solución rápida para generar un modelo predictivo básico antes de considerar modelos más complejos.
2. **Relación lineal entre variables:**
Modelo adecuado para relaciones lineales: Las características (como edad, IMC, fumador, etc.) y los cargos médicos son aproximadamente lineales, por ende la regresión lineal será capaz de modelarlas adecuadamente. Aunque este no siempre sea el caso, la regresión lineal puede servir como un buen punto de partida.
**Características continuas y categóricas:** Dado que las características como "edad", "IMC" y "cargos" son numéricas y las variables como "región" son categóricas, la regresión lineal permite el uso de variables dummy para las categóricas, lo que la hace adecuada para este tipo de problema mixto (variables numéricas y categóricas).

```python

# Convertir 'sex' y 'smoker' a variables binarias
df['sex'] = df['sex'].map({'female': 0, 'male': 1})
df['smoker'] = df['smoker'].map({'no': 0, 'yes': 1})

# Convertir 'region' en variables dummy
df = pd.get_dummies(df, columns=['region'], drop_first=True)

# Separar las características (X) y la variable objetivo (y)
X = df[['age', 'bmi', 'children', 'sex', 'smoker', 'region_southeast', 'region_northwest']]
y = df['charges']

# Escalar las características
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

# Dividir el dataset en entrenamiento y prueba (80% entrenamiento, 20% prueba)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Crear el modelo de regresión lineal
modelo = LinearRegression()

# Entrenar el modelo
modelo.fit(X_train, y_train)

# Hacer predicciones en el conjunto de prueba
y_pred = modelo.predict(X_test)

# Evaluar el modelo
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)  
r2 = r2_score(y_test, y_pred)

# Mostrar las métricas
print(f"Error cuadrático medio (MSE): {mse}")
print(f"Error absoluto medio (MAE): {mae}")
print(f"R² (coeficiente de determinación): {r2}")

# Graficar las predicciones vs los valores reales
plt.scatter(y_test, y_pred)
plt.xlabel("Valores reales")
plt.ylabel("Predicciones")
plt.title("Predicciones vs Valores Reales")
plt.show()
```

```python
# Calcular el error absoluto para cada predicción
errors = np.abs(y_test - y_pred)

# Graficar las predicciones vs los valores reales con colores basados en el error absoluto
plt.figure(figsize=(8, 6))
scatter = plt.scatter(y_test, y_pred, c=errors, cmap='viridis', alpha=0.8, edgecolor='k')
plt.colorbar(scatter, label='Error absoluto')
plt.xlabel("Valores reales")
plt.ylabel("Predicciones")
plt.title("Predicciones vs Valores Reales (coloreado por error)")
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
