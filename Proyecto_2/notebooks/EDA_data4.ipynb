```python
def limpiar_cadena(cadena):
    """
    Limpia una cadena de texto realizando las siguientes operaciones:
    1. Convierte todo el texto a minúsculas.
    2. Elimina caracteres no imprimibles antes de la primera letra y después de la última letra,
       pero mantiene los caracteres internos.
    
    Parámetros:
    - cadena (str): La cadena de texto a limpiar.
    
    Retorna:
    - str: La cadena limpia.
    """
    
    if isinstance(cadena, str):
        # 1. Convertir todo a minúsculas
        cadena = cadena.lower()
        # 2. Eliminar caracteres no imprimibles antes de la primera letra y después de la última letra
        # Buscar la posición de la primera letra (a-z)
        primer_letra = re.search(r'[a-z]', cadena)
        # Buscar la posición de la última letra (a-z)
        ultima_letra = re.search(r'[a-z](?!.*[a-z])', cadena)
        
        if primer_letra and ultima_letra:
            inicio = primer_letra.start()
            fin = ultima_letra.end()
            cadena = cadena[inicio:fin]
        else:
            # Si no se encuentran letras, eliminar espacios en blanco
            cadena = cadena.strip()
        
        return cadena
    return cadena

# Identificar automáticamente las columnas categóricas (tipo object, string o category)
categorical_cols = df.select_dtypes(include=['object', 'string', 'category']).columns

# Aplicar la limpieza a las columnas categóricas
for col in categorical_cols:
    df[col] = df[col].apply(limpiar_cadena)
# Verificar los cambios
for col in categorical_cols:
    print(f"\nValores únicos después de limpieza en '{col}':")
    print(df[col].unique())
```

**Justificación de la Elección del Clustering:**
**Objetivo del Clustering:**

El clustering es adecuado para este conjunto de datos porque no se trata de predecir una etiqueta específica (como en la clasificación), sino de descubrir agrupamientos naturales en los datos sin tener información previa sobre las etiquetas (especies de flores).
El objetivo es identificar patrones y grupos de flores similares basados en sus características (longitud y ancho del sépalo y pétalo), lo que es un problema clásico de clustering.
Relevancia del Clustering en el Contexto:

El clustering es relevante en muchas aplicaciones prácticas, como la segmentación de clientes, la organización de productos y en la biología para agrupar especies similares. Aplicar clustering a un conjunto de datos como Iris puede ayudar a entender mejor las diferencias y similitudes entre las especies de flores basándose en sus características físicas.
Desafíos:

Aunque el dataset Iris es relativamente limpio, el desafío principal será la elección del número correcto de clusters. Aunque sabemos que hay tres especies, el algoritmo de clustering no tiene etiquetas, por lo que necesitamos determinar de manera automática cuántos clusters son necesarios.
La escalabilidad de las características numéricas también es un desafío, ya que las variables tienen diferentes rangos y unidades, lo que podría afectar a algunos algoritmos de clustering como K-means. Por ello, será necesario escalar las características para obtener mejores resultados.
Tipo de Clustering a Utilizar:
Para este caso, se podría utilizar el algoritmo K-means o el Método de Clustering Jerárquico:

K-means: Es un método simple y eficiente para realizar clustering. Se asume que los clusters son esféricos y de tamaño similar.
Clustering Jerárquico: No requiere especificar el número de clusters previamente, lo que podría ser útil para explorar diferentes agrupamientos.
En resumen, el objetivo es aplicar un algoritmo de clustering para agrupar las flores en función de sus características morfológicas, sin utilizar la columna de especies, y evaluar el rendimiento del modelo basándonos en la distribución de las flores en los grupos resultantes.

```python
# Preparar datos para clustering
clustering_data = df.iloc[:, 1:5]
print(clustering_data.head())
```

```python
# Escalar los datos
scaler = StandardScaler()
scaled_data = scaler.fit_transform(clustering_data)

# Convertir los datos escalados en un DataFrame
scaled_df = pd.DataFrame(scaled_data, columns=clustering_data.columns)
scaled_df.head()
```

```python
# Lista para guardar la inercia (distancia intra-clúster)
inertia = []
k_range = range(1, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_data)
    inertia.append(kmeans.inertia_)

# Graficar el método del codo
plt.figure(figsize=(8, 5))
plt.plot(k_range, inertia, marker='o')
plt.xlabel('Número de Clústeres (k)')
plt.ylabel('Inercia')
plt.title('Método del Codo')
plt.show()
```

```python
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_data)
    score = silhouette_score(scaled_data, kmeans.labels_)
    print(f'k={k}, Silhouette Score={score:.3f}')
```

```python
# Entrenar el modelo 
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(scaled_data)

# Agregar etiquetas de clúster al DataFrame original
df['Cluster'] = kmeans.labels_
df.head()
```

```python
# Reducir a 2 dimensiones con PCA
pca = PCA(n_components=2)
pca_data = pca.fit_transform(scaled_data)

# Convertir a DataFrame
pca_df = pd.DataFrame(pca_data, columns=['PC1', 'PC2'])
pca_df['Cluster'] = kmeans.labels_

# Graficar
plt.figure(figsize=(8, 5))
for cluster in pca_df['Cluster'].unique():
    cluster_data = pca_df[pca_df['Cluster'] == cluster]
    plt.scatter(cluster_data['PC1'], cluster_data['PC2'], label=f'Cluster {cluster}')

plt.title('Visualización de Clústeres')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend()
plt.show()
```

```python
cluster_means = df.groupby('Cluster').mean(numeric_only=True)
print(cluster_means)
```

```python
# Creamos un gráfico de dispersión para las características 'sepallengthcm' y 'sepalwidthcm'
sns.scatterplot(data=df, x='sepallengthcm', y='sepalwidthcm', hue=kmeans.labels_, palette='viridis')
plt.title("Distribución de Clusters en función del sépalo")
plt.xlabel("Longitud del Sépalo (cm)")
plt.ylabel("Ancho del Sépalo (cm)")
plt.legend(title="Cluster")
plt.show()
```

```python
silhouette_avg = silhouette_score(df[['sepallengthcm', 'sepalwidthcm', 'petallengthcm', 'petalwidthcm']], kmeans.labels_)
print(f"Índice de silueta: {silhouette_avg}")
```
