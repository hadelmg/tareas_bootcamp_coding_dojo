### 2. Crear Histogramas:
   * Crea histogramas para las variables Score, GDP per capita, Healthy life expectancy y Freedom to make life choices.
   * Personaliza los histogramas con colores, bordes y ajustes de transparencia.


```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Variables a graficar
variables = ['happiness_score', 'economy', 'health', 'freedom']

# Crear histogramas
plt.figure(figsize=(12, 10))

for i, var in enumerate(variables):
    plt.subplot(2, 2, i + 1)
    sns.histplot(df[var], bins=20, color='skyblue', edgecolor='black', alpha=0.7)
    plt.title(f'Histograma de {var.replace("_", " ").title()}', fontsize=14)
    plt.xlabel(var.replace("_", " ").title())
    plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```

La distribución de los puntajes de felicidad de los países parece ser aproximadamente normal, con una ligera asimetría hacia la derecha, lo que indica que la mayoría de los países tienen un puntaje moderado (entre 4 y 6), aunque algunos con puntajes altos desplazan la media ligeramente hacia la derecha. El pico de la distribución se encuentra en torno a 5.5, lo cual sugiere que es el puntaje más común, y la dispersión de los datos muestra una considerable variabilidad en los niveles de felicidad entre países. Aunque no hay valores atípicos extremos, se observan algunos puntajes muy bajos (cerca de 3) y otros bastante altos (cerca de 7).


----

La distribución de los valores de Economy en el conjunto de datos es ligeramente sesgada hacia la derecha, con la mayoría de los datos concentrados en la parte izquierda del histograma en valores más bajos y una cola más larga hacia la derecha. El valor más común se encuentra alrededor de 1, y aunque existe una dispersión moderada que indica una variedad de valores, la mayoría están cerca de este centro. No se observan valores atípicos extremos, lo cual sugiere que los datos son relativamente homogéneos.

La distribución de los valores de Health en el conjunto de datos muestra una ligera inclinación hacia la derecha, con la mayoría de los datos concentrados en valores más bajos a la izquierda del histograma y una cola más larga hacia la derecha. El valor más común es alrededor de 0.8, y aunque hay una dispersión moderada que indica una variedad de valores, la mayoría se encuentra cerca de este punto central. La distribución de los valores de Freedom en el conjunto de datos muestra una ligera inclinación hacia la derecha, con la mayoría de los datos concentrados en valores bajos a la izquierda del histograma y una cola más larga hacia la derecha. ### 3. Crear Boxplots:
   * Crea boxplots para las mismas variables para visualizar la mediana, el rango intercuartil y los valores atípicos.
   * Personaliza los boxplots con diferentes colores y estilos.


```python
# Crear boxplots
plt.figure(figsize=(12, 10))

for i, var in enumerate(variables):
    plt.subplot(2, 2, i + 1)
    sns.boxplot(data=df, y=var, color='lightcoral')
    plt.title(f'Boxplot de {var}', fontsize=14)

plt.tight_layout()
plt.show()
```

El rango intercuartílico del puntaje de felicidad abarca de aproximadamente 4 a 6.5, con una mediana cercana a 5.5, lo que indica que la mitad de los países tiene puntajes por encima de 5.5 y la otra mitad por debajo. Se observan valores atípicos tanto altos como bajos, donde los puntajes superiores reflejan países con niveles de felicidad excepcionalmente altos, y los inferiores, lo contrario. La distribución presenta una ligera asimetría positiva, con una "cola" superior más larga, sugiriendo que algunos países con puntajes altos están sesgando la distribución hacia arriba.

----

El rango intercuartílico de "economy" se extiende de aproximadamente 0.5 a 1.25, lo que indica que la mayoría de los países tienen valores en este rango, con una mediana cercana a 1, dividiendo a los países en aquellos con valores de "economy" por encima o por debajo de 1. Los valores atípicos superiores reflejan países con valores de "economy" excepcionalmente altos, y la distribución muestra una ligera asimetría positiva, con una "cola" superior más larga, sugiriendo que algunos países con valores altos están sesgando la distribución hacia arriba.

----

El rango intercuartílico de los datos de salud se extiende de aproximadamente 0.4 a 0.8, indicando que la mayoría de los individuos tienen niveles de salud dentro de este intervalo. La mediana, situada en 0.7, muestra que la mitad de los individuos tienen niveles de salud por encima de este valor y la otra mitad por debajo. Existen valores atípicos tanto superiores como inferiores, con los primeros reflejando individuos con niveles de salud excepcionalmente altos y los inferiores, niveles muy bajos. La distribución es ligeramente asimétrica positiva, con una cola superior más larga que sugiere que algunos valores altos están desplazando la distribución hacia arriba.

----

El rango intercuartílico de "libertad" va de aproximadamente 0.3 a 0.6, indicando que la mayoría de los individuos tienen niveles de libertad dentro de este rango, con una mediana cercana a 0.5 que divide a los individuos en aquellos con niveles de libertad por encima o por debajo de este punto. ### 4. Análisis Combinado:
   * Combina los histogramas y los boxplots en una sola figura para cada variable.
   * Interpreta los resultados y escribe un análisis sobre la distribución de cada variable y la presencia de valores atípicos.


```python
# Análisis combinado: histogramas y boxplots
plt.figure(figsize=(12, 10))

for i, var in enumerate(variables):
    plt.subplot(2, 2, i + 1)
    sns.histplot(df[var], bins=20, color='skyblue', edgecolor='black', alpha=0.7, label='Histograma')
    sns.boxplot(data=df, y=var, color='lightcoral', width=0.2)
    plt.title(f'Histograma y Boxplot de {var}', fontsize=14)
    plt.xlabel(var)
    plt.ylabel('Frecuencia')
    plt.legend()

plt.tight_layout()
plt.show()
```

Tanto el histograma como el boxplot indican que la mayoría de los países tienen un nivel de felicidad moderado, pero existe una considerable variabilidad entre ellos. Hay algunos países con niveles de felicidad excepcionalmente altos y otros con niveles muy bajos. La ligera asimetría positiva sugiere que hay algunos países con puntajes de felicidad muy altos que están "jalando" la distribución hacia arriba.
