# Análisis de Ventas y Predicción de Comportamiento de Clientes

## 1. Introducción y Objetivos del Proyecto
### Objetivos:
- Preprocesar y analizar un conjunto de datos de ventas.
- Implementar múltiples modelos de machine learning para predecir el comportamiento de compra (`total_amount`).
- Comparar el rendimiento de diferentes modelos.
- Seleccionar el modelo más adecuado para el análisis de ventas y comportamiento de clientes.

---

## 2. Descripción del Conjunto de Datos
El conjunto de datos contiene información sobre transacciones de ventas, con columnas como:

- `transaction_id`: Identificador único de la transacción.
- `customer_id`: Identificador único del cliente.
- `gender`: Género del cliente.
- `age`: Edad del cliente.
- `product_category`: Categoría del producto.
- `quantity`: Cantidad de productos vendidos.
- `price_per_unit`: Precio por unidad de producto.
- `total_amount`: Monto total de la transacción.

Objetivo de la predicción: Predecir el `total_amount`

## 3. Principales Análisis y Hallazgos

### Análisis Exploratorio de Datos:
- Se verificaron los valores faltantes y la distribución de las variables.
- Se realizaron transformaciones para asegurar que los datos estén listos para el modelado:
  - Codificación de variables categóricas (`gender`, `product_category`).
  - Escalado de variables numéricas (`age`, `quantity`, `price_per_unit`).

### Modelos Evaluados:
Se compararon los siguientes modelos de regresión para predecir el `total_amount`:

1. **Regresión Lineal (Ridge)**
2. **K-Nearest Neighbors (KNN)**
3. **Árbol de Decisión**
4. **Random Forest**
5. **XGBoost**
6. **LightGBM**
